```python
import imageio.v2 as imageio
import matplotlib as mpl
import matplotlib.pyplot as plt
import numpy as np
from IPython.display import Video
from matplotlib.colors import CenteredNorm
from openpmd_viewer import OpenPMDTimeSeries
```

```python
series = OpenPMDTimeSeries("diags/diag1/")
iterations = series.iterations
```

```python
i = iterations[-1]
z1, uz1 = series.get_particle(["z", "uz"], species="electrons1", iteration=i)
z2, uz2 = series.get_particle(["z", "uz"], species="electrons2", iteration=i)
plt.scatter(z1, uz1)
plt.scatter(z2, uz2)
plt.title(f"{i}")
plt.show()
```

```python
series = OpenPMDTimeSeries(f"diags/diag1")
iterations = series.iterations

video_filename = "phase_space.mp4"

with imageio.get_writer(video_filename, fps=6, codec="libx264") as writer:
    for i in iterations:

        fig, ax = plt.subplots(figsize=(6, 4))
        z1, uz1 = series.get_particle(["z", "uz"], species="electrons1", iteration=i)
        ax.scatter(z1, uz1)
        z2, uz2 = series.get_particle(["z", "uz"], species="electrons2", iteration=i)
        ax.scatter(z2, uz2)
        ax.set_xlabel("z")
        ax.set_ylabel("uz")
        ax.set_title(f"Iteration {i}")
        ax.grid(True)

        fig.tight_layout()
        fig.canvas.draw()
        image = np.asarray(fig.canvas.buffer_rgba())[:, :, :3]
        writer.append_data(image)
        plt.close(fig)

Video(video_filename, embed=True)
```
